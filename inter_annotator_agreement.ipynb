```python
import pandas as pd
from sklearn.metrics import cohen_kappa_score,confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Read the CSV File
#Change accordingly to ur workspace
annotator1 = pd.read_csv("C:\\Users\\User\\Desktop\\popular_comment_Bolt_zh_annotate.csv")
annotator2 = pd.read_csv("C:\\Users\\User\\Desktop\\popular_comment_Bolt_YWAnnotate.csv")
```

```python
annotator2
```

```python
#Joining the data
annotator1['annotator 2'] = annotator2['annotator 1']
annotator1['irony 2'] = annotator2['irony 1']
annotator1['remove 2'] = annotator2['remove 1']

annotator1.to_csv("popular_comment_Bolt_annotate_Merged.csv", index = True)
```

```python
annotator1
```

```python
#Calculate Inter-Annotator-Agreement
kappa_annotator = cohen_kappa_score(annotator1['annotator 1'],annotator1['annotator 2'])
kappa_irony = cohen_kappa_score(annotator1['irony 1'],annotator1['irony 2'])

print(f"\n Cohen's Kappa Score between Annotator : {kappa_annotator}")
print(f"\n Kappa Score for Irony: {kappa_irony}")
```

```python
#Confusion Matrix
cm = confusion_matrix(annotator1['annotator 1'],annotator1['annotator 2'])

print(f"\n Confusion Matrix between Annotator")
print(cm)

plt.figure(figsize = (10,7))
sns.heatmap(cm,annot = True, fmt = 'g')
plt.xlabel('Annotator 1')
plt.ylabel('Annotator 2')
plt.title('Confusion Matrix between Annotator')
plt.show()
```
